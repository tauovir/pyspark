#### Store the dictinary by key order

```python
from collections import OrderedDict

dict1 = {"2":"mango","5":"orange","1":"date"}
print(dict1)
dict_values = [val for val in dict1.keys()]
ord_dict = OrderedDict()
dict_values.sort()
for ele in dict_values:
    ord_dict.update({ele:dict1[ele]})

print(ord_dict)

```

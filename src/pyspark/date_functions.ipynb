```python
from pyspark.sql import functions as F
```

```python
%sql
CREATE OR REPLACE TEMP VIEW dataframe_view AS
SELECT * FROM zomato LIMIT 5
```

```python
%sql
select order_place_time::date ,
date_add(order_place_time,2) as date_added -- date_add(start_date, num_days) - Returns the date that is num_days
,date_diff(date('2024-01-20'),order_place_time) as date_difference  -- Returns the number of days 
,date_format(order_place_time, 'y') as format_year
,date_format(order_place_time, 'D') as format_day -- check https://spark.apache.org/docs/latest/sql-ref-datetime-pattern.html
,date_from_unix_date(2) as unix_date -- Create date from the number of days since 1970-01-01.
,date_part('minute',order_place_time) as date_part_minute
-- ,date_trunc('MINUTE',order_place_time) as date_truncate_minute
from zomato limit 5;
```

```python
(
    spark.table('dataframe_view')
    .select('order_place_time'
            ,F.to_date(F.col('order_place_time'))
            ,F.date_add(F.col('order_place_time'),2).alias('date_added')
            ,F.date_format(F.col('order_place_time'),'y').alias('format_year')
            ,F.date_from_unix_date(F.lit(40)).alias('unix_date')
            ,F.date_part(F.lit('minute'),F.col('order_place_time')).alias('unix_date')
            )
 
).display()
```
